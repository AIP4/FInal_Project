```python
import torch
import torch.nn as nn

# 예제 데이터: shape (m, n)
m, n = 10, 5
data = torch.randn(m, n)

# Conv1D 입력 형태로 변환: (batch_size, seq_length, num_features)
data = data.unsqueeze(0)  # (1, m, n)
```

```python
data
```

```python
conv1d = nn.Conv1d(in_channels=n, out_channels=16, kernel_size=3, padding=1)
output = conv1d(data)
conv1d2 = nn.Conv1d(in_channels=16, out_channels=16, kernel_size=3, padding=1)
output2 = conv1d2(output)
```

```python
print(output.shape)
print(output2.shape)
```

```python
import einops

outputs = []
outputs.append(output)
outputs.append(output2)

outputs = torch.stack(outputs, dim=1)
print(outputs.shape)
```

```python


arranged = einops.rearrange(outputs, 'b k c seq -> b seq (k c)')
print(arranged.shape)
```

```python
print(arranged)
```

```python
output = einops.rearrange(output, 'b c seq -> b seq c')
output
```

```python
print(output)
print(output2)
```

```python
from einops import rearrange

# 변환된 코드
data = rearrange(data, 'b seq feats -> b feats seq')
output = conv1d(data)
print(output.shape)
print(output)
```
