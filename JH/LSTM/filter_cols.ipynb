```python
import pandas as pd
from utils import prepare_data
from interpolate import simple_interpolate
from dataset import FinedustDataset

region = "Andong"
df = prepare_data(region.lower())
df = simple_interpolate(df)
dataset = FinedustDataset(df)
```

```python
df
```

```python
correlation_matrix = df[dataset.feature_columns].corr()

# 상관관계가 높은 feature 선택 (예: 0.5 이상의 상관관계)
significant_features = correlation_matrix[correlation_matrix.abs() > 0.5].index.tolist()

# 유의미한 feature로 새로운 데이터프레임 생성
df_significant = df[significant_features]

# 결과 출력
print(df_significant)
```
