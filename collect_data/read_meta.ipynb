```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
meta = pd.read_csv("./data/kma/META_관측지점정보.csv", encoding="cp949")
meta.head()
```

```python
meta["지점명"].unique()
```

```python
import folium

map_center = [meta['위도'].mean(), meta['경도'].mean()]
m = folium.Map(location=map_center, zoom_start=10)

for _, row in meta.iterrows():
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=f"{row['지점명']} ({row['지점']})",
        tooltip=row['지점명'],
    ).add_to(m)

m.save("korea_weather_stations_map.html")
m
```

* 지점 선택
* [`제주 (184)`, `서울 (108)`, `춘천 (101)`, `대전 (133)`, `안동 (136)`, `전주 (146)`, `대구 (143)`, `광주 (156)`, `부산 (159)`]
* 위도/경도에 따라 9개 지역 선택

* 관측 종료 지점 제거

```python
### Filtered folium (9 locations)

# 관측 종료 지점 제거
meta_current = meta[meta["종료일"].isnull()]
locations = [184, 108, 101, 133, 136, 146, 143, 156, 159]
meta_filtered = meta_current[meta_current["지점"].isin(locations)]

map_center = [meta_filtered['위도'].mean(), meta_filtered['경도'].mean()]
m = folium.Map(location=map_center, zoom_start=10)

for _, row in meta_filtered.iterrows():
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=f"{row['지점명']} ({row['지점']})",
        tooltip=row['지점명'],
    ).add_to(m)

m.save("korea_weather_stations_map.html")
m
```
