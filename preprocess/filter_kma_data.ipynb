```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../collect_data/data/kma/kma_2022.csv")
df.head()
```

< Description >
* -9: Missing value

< Goal >
* Extract meaningful variables

# Check missing value rate

```python
df_replaced = df.replace(-9, np.nan)
missing_ratio = df_replaced.isnull().mean() * 100

print("Missing ratio (%)")
print(missing_ratio)
```

```python
# 결측치 비율이 20%를 넘는 열 제거
df_filtered = df_replaced.loc[:, missing_ratio <= 20]
df_filtered.shape
```

```python
# Recalculate missing ratio
missing_ratio = df_filtered.isnull().mean() * 100

print("Missing ratio (%)")
print(missing_ratio)
```

```python
import os
os.makedirs("../collect_data/filtered/kma")

### Save dataframe
df_filtered.to_csv("../collect_data/filtered/kma/kma_2022.csv", index=False)
```
